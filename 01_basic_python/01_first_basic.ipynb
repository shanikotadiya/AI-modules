```python
users = [
    {"name": "Shani", "role": "MERN", "score": 85},
    {"name": "Raj", "role": "Backend", "score": 92},
    {"name": "Meera", "role": "Frontend", "score": 78}
]
score = 0
for user in users:
    if(user["score"] > score):
        score = user["score"]
    if(user["name"] == "Meera"):
        user["score"] = 88
users.append({
    "name":"riya",
    "role": "teacher",
    "score": 25
})
print(score)
print(users)
```

```python
mixed_data = ["10", 20, "30.5", 40, "fifty", True, "60"]

newdata = []
for i in mixed_data:
    if type(i) == bool:
        continue
    try:
        newdata.append(float(i))
    except:
        continue
print(newdata)
```
