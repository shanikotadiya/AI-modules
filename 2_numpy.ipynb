```python
import numpy as np

marks = np.array([45, 60, 72, 88, 54, 90])

print("Mean:", np.mean(marks))
print("Max:", np.max(marks))
print("Min:", np.min(marks))
print("Pass/Fail:", marks >= 60)
print("Grace Marks:", marks + 5)
```

```python
data = ["12", "15.5", 20, "AI", None, True, "30"]

clearnumber = [
    int(item)
    for item in data
    if isinstance (item, int) and not isinstance(item, bool)
    or (isinstance(item, str) and item.isdigit())
]

print(clearnumber)
totalnumber = sum(clearnumber)
print(totalnumber / len(clearnumber))
```
